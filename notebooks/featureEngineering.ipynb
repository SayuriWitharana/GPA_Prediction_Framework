```python
from pathlib import Path
import pandas as pd

DATA_PATH = Path("../data/raw/TotalDataSet.xlsx") 
df = pd.read_excel(DATA_PATH)

def assign_perf_group(gpa):
    if gpa <= 2.99:
        return "Underperforming"
    elif gpa <= 3.29:
        return "Average"
    else:
        return "Performing"

df["GroupLabel"] = df["FinalGPA"].apply(assign_perf_group)

# Check distribution
print(df["GroupLabel"].value_counts())
```

Performance group labels were derived using fixed GPA thresholds based on institutional grading standards. As this step does not involve parameter estimation from data, it was performed prior to cross-validation and persisted for reproducibility.
